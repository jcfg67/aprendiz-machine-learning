# Linear Classification

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Visualization Style
%matplotlib inline
#plt.style.use('seaborn-white')
plt.style.use('fivethirtyeight')
plt.tight_layout()
```

### Create Dataset

```python
X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)
X.shape, y.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="autumn")
plt.show()
```

## Support Vector Machine

```python
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
#    if plot_support:
#        ax.scatter(model.support_vectors_[:, 0],
#                   model.support_vectors_[:, 1],
#                   s=300, linewidth=1, edgecolor='gray',
#                   facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model)
```

```python
model.support_vectors_
```

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

```python
from ipywidgets import interact, fixed
interact(plot_svm, N=[10, 50, 100, 200], ax=fixed(None));
```

## Example: Adult Income Dataset

```python
data = pd.read_csv("./datasets/adult.csv", index_col=0)
data.head()
```

```python
income = data.income
data_features = data.drop("income", axis=1)
```

```python
display(data_features.head())
```

```python
data.drop(columns = ['workclass', 'education', 'marital-status', 
                     'occupation', 'relationship', 'capital-gain', 
                     'capital-loss', 'native-country'], inplace=True)
```

#### Data Encoding: using dummies

```python
dummies = pd.get_dummies(data.income)
dummies.head()
data = pd.concat([data,dummies], axis='columns')
data.head()
```

```python
dummies = pd.get_dummies(data.race)
dummies.head()
data = pd.concat([data,dummies], axis='columns')
data.head()
```

```python
dummies = pd.get_dummies(data.gender)
dummies.head()
data = pd.concat([data,dummies], axis='columns')
data.head()
```

```python
data.drop(columns = ['race', 'gender'], inplace=True)
data.head()
```

```python
### Preprocessing

X_train, X_test, y_train, y_test = train_test_split(data, income)

#scaler = StandardScaler().fit(X_train)
#X_train_scaled = scaler.transform(X_train)
```

```python
### Cross-validation with default parameters
#scores = cross_val_score(LogisticRegression(), X_train, y_train, cv=5)
#print(scores.mean())
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
### do grid search

param_grid = {'C': np.logspace(-3, 3, 7)}
param_grid


grid = GridSearchCV(LogisticRegression(solver='lbfgs'), param_grid, cv=5,
                    return_train_score=True)

grid.fit(X_train, y_train)

grid.best_params_
grid.best_score_
```

```python
# some visualization

res = pd.DataFrame(grid.cv_results_)
res.mean_test_score.plot()
res.mean_train_score.plot()
plt.xscale("log")

grid.score(X_test, y_test)

important = np.argsort(np.abs(grid.best_estimator_.coef_)).ravel()

plt.barh(range(10), grid.best_estimator_.coef_.ravel()[important[-10:]])
plt.yticks(range(10), X_train.columns[important[-10:]]);
```
