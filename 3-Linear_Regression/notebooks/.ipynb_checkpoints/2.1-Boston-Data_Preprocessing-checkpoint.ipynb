# Data Preprocessing

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

## Data Loading

```python
boston = datasets.load_boston()
dir(boston)
```

```python
print(boston.DESCR)
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df.head()
```

```python
df.boxplot(rot=45)
```

## Data Exploring

```python
X, y = boston.data, boston.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## Data Visualization

```python
fig, axes = plt.subplots(3, 5, figsize=(20, 10))
for i, ax in enumerate(axes.ravel()):
    if i > 12:
        ax.set_visible(False)
        continue
    ax.plot(X[:, i], y, 'o', alpha=.5)
    ax.set_title("{}: {}".format(i, boston.feature_names[i]))
    ax.set_ylabel("MEDV")
```

```python
plt.boxplot(X)
plt.xticks(np.arange(1, X.shape[1] + 1),
           boston.feature_names, rotation=30, ha="right");
plt.savefig("../images/boston_unscaled_box.png")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
plt.boxplot(X_train_scaled)
plt.xticks(np.arange(1, X.shape[1] + 1),
           boston.feature_names, rotation=30, ha="right");
plt.savefig("../images/boston_scaled_box.png")
```

```python
knr = KNeighborsRegressor().fit(X_train, y_train)
knr.score(X_train, y_train)
```

```python
knr.score(X_test, y_test)
```

```python
knr_scaled = KNeighborsRegressor().fit(X_train_scaled, y_train)
knr_scaled.fit(X_train_scaled, y_train)
knr_scaled.score(X_train_scaled, y_train)
```

```python
X_test_scaled = scaler.transform(X_test)
knr_scaled.score(X_test_scaled, y_test)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
rf_scaled = RandomForestRegressor(n_estimators=100, random_state=0)
rf_scaled.fit(X_train_scaled, y_train)
rf_scaled.score(X_test_scaled, y_test)
```

# Categorical Variables

```python
df = pd.DataFrame({'salary': [103, 89, 142, 54, 63, 219],
                   'boro': ['Manhattan', 'Queens', 'Manhattan', 'Brooklyn', 'Brooklyn', 'Bronx']})
df
```

```python
pd.get_dummies(df)
```

```python
categorical = df.dtypes == object
categorical
```

```python
ct = make_column_transformer((OneHotEncoder(), categorical),
                             (StandardScaler(), ~categorical))
ct.fit_transform(df)
```
